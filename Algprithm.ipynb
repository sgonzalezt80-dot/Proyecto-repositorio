# Algorithm that delete duplicate numbers

```python
import random 
def delete_duplicate(lista):                #1
    i = 0                                   #1

    while i < len(lista):                   #n
        j = i + 1                           #n(1)

        while j < len(lista):               #n * n 
            
            if lista[i] == lista[j]:        #n * n 

                lista.pop(j)                #n * n (Como no podemos usar n/2, lo tomamos como n * n)

            else:                           #n * n
                j = j + 1                   #n * n
        i = i + 1                           #n
    
    return lista                            #1
```

##### Prueba de que el algoritmo funciona 

```python
def delete_duplicate_empiric():
    my_list = []
    n = 10
    i = 1
    while(i < n):
        my_list.append(random.randint(1, 9))
        i += 1

    print(my_list)
    print(delete_duplicate(my_list))

delete_duplicate_empiric()
```

# Graph

```python
import matplotlib.pyplot as plt 
```

```python
def time_delete_duplicate(lista):
    i = 0
    times = 0
    while i < len(lista):
        j = i + 1
        times += 1
        while j < len(lista):
            times += 1
            if lista[i] == lista[j]:
                #times += 1
                lista.pop(j)

            else:
                j = j + 1
        i = i + 1
    
    return times

def graph_delete_duplicate():
    x = []
    y = []

    size = 5
    MAX = 25

    while(size < MAX):
        i = 0
        ml = []

        while(i < size):
            ml.append(random.randint(1, 2000))
            i = i + 1

        x.append(size)
        y.append(time_delete_duplicate(ml))

        size = size + 1
    plt.plot(x, y)

graph_delete_duplicate()
```

```python
def graph_delete_duplicate_worst():
    x = []
    y = []

    size = 5 #tamaño de la primer lista 
    MAX = 50

    while(size < MAX):
        ml = []
        i = 0
        j = 1
        while(i < size):
            ml.append(j)
            i = i + 1
            j = j + 1

        x.append(j)
        y.append(time_delete_duplicate(ml)) 

        size = size + 1
    plt.plot(x, y)

graph_delete_duplicate_worst()
```

```python
def graph_delete_duplicate_best():
    x = []
    y = []

    size = 5 #tamaño de la primer lista 
    MAX = 50

    while(size < MAX):
        ml = []
        i = 0
        
        while(i < size): 
            ml.append(1)
            i = i + 1

        x.append(i)
        y.append(time_delete_duplicate(ml))
        size = size + 1
    plt.plot(x, y)

graph_delete_duplicate_best()
```
